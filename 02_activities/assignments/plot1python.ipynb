```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("vital_events_data_by_month_1994-2024.csv")

# Create a mapping of month names to numbers 
month_map = {
    'January/janvier': 1, 'February/février': 2, 'March/mars': 3,
    'April/avril': 4, 'May/mai': 5, 'June/juin': 6,
    'July/juillet': 7, 'August/août': 8, 'September/septembre': 9,
    'October/octobre': 10, 'November/novembre': 11, 'December/décembre': 12
}

# Map month names to numbers
df['Month_Num'] = df['Month/Mois'].map(month_map)

# datetime
df['Date'] = pd.to_datetime(df['Year/Année'].astype(str) + '-' + df['Month_Num'].astype(str) + '-01')

# Date as index
df.set_index('Date', inplace=True)

# target columns
df_plot = df[['Live Births/Naissances vivantes', 'Marriages/Mariages', 'Deaths/Dècés']]

# Check the first few rows
print(df_plot.head())

# Plot
plt.figure(figsize=(14,6))
sns.lineplot(data=df_plot)
plt.title("Population Events Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(title='Event')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
