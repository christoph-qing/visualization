### Choose the Right Visualization

Figure is the whole thing that python generate, axis is one part of the figure, for example, the region within x, y, axis

There are visualizations that are meant to deliver an emotional response and there are more neutral ones that present data.

Some are meant to show all are above a threshold, some can show an overal trend.
Is the visualization pleasing to look at? → Aesthetic
Does the visualization accurately and honestly present data? → Substantive
Can we understand what message the maker of the visualization is attempting to convey? → Perceptual

Tables, networks, fields, and geometries are four basic dataset types.
