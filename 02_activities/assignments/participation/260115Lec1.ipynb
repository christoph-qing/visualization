```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

What we see is a figure, the plot is the axes, not x/y axis

```python
# Generate random data
np.random.seed(1)
# from 0 to 49, integer
x = np.arange(50)
# generate 50 random integer with value between 0 and 100
y = np.random.randint(0, 100, 50)
```

```python
# length 5 width 3
fig, ax = plt.subplots(figsize=(5,3))
# use both random number x and y to set up the scattered plot
ax.scatter(x,y)
```

```python
# Use bar plot instead
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
#line plot
# Use bar plot instead
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# Histogram
# Use bar plot instead
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) # We don't need to add the x value because we are showing the frequency of y

# Add boundaries for plots
ax.hist(y, edgecolor = "black")
```

Plot Labels

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
# Add to the line plot labels and tatles
ax.set_title('Total growth over time')
ax.set_ylabel('Toal growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

Modifying labels and titles

```python
# Set font
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Toal growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

Modify data points

```python
# Set up colour and marker for data points

fig, ax = plt.subplots(figsize = (5, 3))

ax.scatter(x, y, marker = '*', color = 'indigo')

fig.show()
```

Adjust line and connection

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, marker = '*', color = 'indigo', linestyle = '--', linewidth = 2)

# we can name the color or use hex code
ax.plot(x, y, marker = '*', color = '#7425b9', linestyle = '--', linewidth = 2, markeredgecolor = '#fa9359', markerfacecolor = '#000000')
fig.show()
```

Grid lines

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, marker = '*', color = 'indigo', linestyle = '--', linewidth = 2)
ax.grid(axis = 'y', color = 'blue', linewidth = 2, linestyle = '-.')

fig.show()
```
