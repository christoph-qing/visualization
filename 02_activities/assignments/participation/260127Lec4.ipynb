# Subplots and combining visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) # There are more advanced operations possible
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# two plots next to each other, so we just have to define multiple axes and their relative positions:
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3)) # specify both plots together
```

```python
# Define types of visuals
fig, (ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
ax1.scatter(x1,y1) # axis 1 is scatter plot
ax2.bar(x2,y2) # axis 2 is a bar plot
fig.show()
```

Customize plots

```python
# Arrange subplots within a figure
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']], # This is the layout of first and second row layout, so ax3 will occupy the right side of both ax1 and 2
figsize=(7, 4))
```

```python
# We can add data, and choose different plots for each axis
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
# Lay out adding some very large x axis titles to our previous plot
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

The axis title for our scatter plot
doesnt appear, and the axis title
for our line plot is cut off

We can use layouts to make sure that our subplots fit neatly in our figure area
There are two main kinds of layout we consider:
Tight layout adjusts subplots so tick labels, axis labels, and titles don't overlap or
leave the figure area
Constrained layout works similarly except it also fits things like legends or
colorbars

```python
# Lay out adding some very large x axis titles to our previous plot
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4), layout = 'constrained')
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
# fig.tight_layout() # This does not adjust legends and colour bar
```

```python
# Multiple vis on one axis object

#call multiple plot methods on the same axes

# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

```python
# Add error bar
y2_sd = np.std(y2) # calculate SD
fig, ax = plt.subplots(figsize=(7, 3))
# ax.plot(x, y2, color = "red") # plotline

# Then use errorbar() to add in our error line (standard deviation in this case, but could be whatever value you calculated)
# yerr specifies that were plotting vertical error bars
# fmt makes sure were not plotting the actual data points, only the error

ax.errorbar(x, #our x values
y2, #our y values
color = 'red',
yerr=y2_sd,
fmt = "none") # Can change different fmt
```

```python
# Customize error bar
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")
ax.errorbar(x,
y2,
yerr = y2_sd,
fmt = "none",
ecolor= "indigo",
elinewidth= 4,
capsize = 6,
capthick= 4 )
```

```python
# If we dont want to see error bars for every single point, we can specify intervals using errorevery
ax.errorbar(x,
y2,
yerr = y2_sd,
fmt = "none",
ecolor = "indigo",
elinewidth = 4,
capsize = 6,
capthick = 4,
errorevery= 2 )
```
